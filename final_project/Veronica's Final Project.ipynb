```python
# Importing packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading the csv files

circuits = pd.read_csv("data_raw/circuits.csv")
drivers = pd.read_csv("data_raw/drivers.csv")
driver_standings = pd.read_csv("data_raw/driver_standings.csv")
races = pd.read_csv("data_raw/races.csv")
```

```python
display(drivers)
```

```python
display(driver_standings)
```

```python
# cleaning the circuits df

circuits.dtypes

# checking whether each row is numeric or not
circuits["alt"].str.isnumeric()

circuits['location'] = circuits['location'].str.lower()

# renaming country variable
circuits.rename(columns = {'country':'circuit_country'}, inplace = True)
circuits.rename(columns = {'name':'circuit_name'}, inplace = True)
circuits.rename(columns = {'location':'circuit_location'}, inplace = True)

# extracting list of non-numeric values
subset = circuits.query("alt.str.isnumeric() == False")
list_unique = pd.unique(subset["alt"])

list_old = ['\\N','-7']
list_new = [np.nan, -7]

# replacing NaN values
circuits["alt"] = circuits["alt"].replace(list_old, list_new)

# these are the same values
circuits["circuit_country"] = circuits["circuit_country"].replace("united states","usa")

# extracting a clean df
circuits.to_csv("data_clean/circuits.csv")

```

```python
# displaying the clean circuits df
display(circuits)
```

```python
# isolating a list of countries to merge with the drivers df

subset = circuits.query("circuit_country.str.isnumeric() == False")
list_unique = pd.unique(subset["circuit_country"])
print(list_unique)
```

```python
# cleaning the drivers df

drivers.dtypes

# cleaning the nationality column

drivers['nationality'] = drivers['nationality'].str.lower()

drivers['nationality'] = drivers['nationality'].replace("swiss","swedish")
drivers['nationality'] = drivers['nationality'].replace("east german","german")
# this is the same value

subset = drivers[drivers['nationality'] == "american-italian"]
# research for clarification

drivers['nationality'] = drivers['nationality'].replace("argentine-italian","italian")
drivers['nationality'] = drivers['nationality'].replace("american-italian","italian")
# replacing the values of nationality with the country represented by the racer

subset = drivers.query("nationality.str.isnumeric() == False")
list_nationality = pd.unique(subset["nationality"])
print(list_nationality)
# isolating a list of nationalities to replace with the correponding country 

list_country = ['UK','Germany','Spain','Finland','Japan','France','Poland','Brazil',
                'Italy','Australia','Austria','USA','Netherlands','Colombia','Portugal',
                'Canada','India','Hungary','Ireland','Denmark','Argentina','Czech Republic',
                'Malaysia','Switzerland','Belgium','Monaco','Venezuela','New Zealand','Chile',
                'Mexico','South Africa','Liechtenstein','Rhodesia','Uruguay','Thailand',
                'Russia','Indonesia','China']

# creating the country column to merge with drivers
drivers["country_represent"] = drivers["nationality"].replace(list_nationality, list_country)

# cleaning NaN values

drivers["number"].str.isnumeric()

number_subset = drivers.query("number.str.isnumeric() == False")
number_unique = pd.unique(number_subset["number"])

list_old = ['\\N']
list_new = [np.nan]

# replacing the NaN values
drivers["number"] = drivers["number"].replace(list_old, list_new)

# converting number into a numeric column
drivers["number"] = pd.to_numeric(drivers["number"])

# delete nationality column
# we're looking at the countries represented by the racers
drivers = drivers.drop(["nationality"], axis=1)
drivers = drivers.drop(["dob"], axis=1)
drivers = drivers.drop(["url"], axis=1)

# extracting a clean df
drivers.to_csv("data_clean/drivers.csv")

```

```python
# displaying the clean drivers df
display(drivers)
```

```python
# creating the drivers_merge df

# merge drivers df w/ the driver_standings
drivers_merge = pd.merge(drivers,
                       driver_standings[["driverStandingsId","raceId",
                                         "driverId","points","position","wins"]],
                       on = "driverId",
                       how = "left")

# display the final merge
display(drivers_merge)
```

```python
# merge races w/ circuits

circuits_merge = pd.merge(circuits,
                       races[["raceId","circuitId"]],
                       on = "circuitId",
                       how = "left")

#display the final merge
display(circuits_merge)
```

```python
# final marge drivers_merge w/ circuits_merge

driver_circuit_merge = pd.merge(drivers_merge,
                       circuits_merge[["circuitId","circuit_name","circuit_location","circuit_country","raceId"]],
                       on = "raceId",
                       how = "left")

# combining forename and surname to make a variable containing the fullname
driver_circuit_merge["fullname"] = driver_circuit_merge["forename"] + " " + driver_circuit_merge["surname"]

# display the final merge
display(driver_circuit_merge)
```

```python
# summary statistics for each racer

driver_circuit_agg = (driver_circuit_merge.groupby(  ["driverId","fullname","circuit_country"])
                       .agg(mean_points = ('points','mean'),
                            sd_points =   ('points','std'),
                            min_points =  ('points','min'),
                            max_points =  ('points','max'),
                            count_wins   = ('wins',len)))

display(driver_circuit_agg)

# extract the final df
driver_circuit_agg.to_csv("data_clean/driver_circuit_agg.csv")

# Nyck de Vries was kicked out because he was not good enough
```
